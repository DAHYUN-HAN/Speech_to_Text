```python
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
import librosa.display
import numpy as np
import shutil
import pandas as pd
from pandas import DataFrame as df
import csv

import os
import IPython.display
```

```python
def load(file_name):
    wav = file_name

    file_sr = librosa.get_samplerate(wav)
    y, sr = librosa.load(wav, sr=file_sr)
    return y, sr
```
