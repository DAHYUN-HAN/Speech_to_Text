```python
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
import librosa.display
import numpy as np
import shutil
import pandas as pd
from pandas import DataFrame as df
import csv

import os
import IPython.display
```

```python
def load(file_name):
    wav = file_name

    file_sr = librosa.get_samplerate(wav)
    y, sr = librosa.load(wav, sr=file_sr)
    return y, sr
```

```python
y, sr = load('blank_ver_계세원_1 - 복사본 (4).wav')
IPython.display.Audio(data=y[int(86.8*sr):int(105.8*sr)], rate=sr)
```
